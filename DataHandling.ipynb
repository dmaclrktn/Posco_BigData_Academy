### 실습 5

```python
# STEP1 : 데이터 구성하기

df4 = pd.DataFrame({"A" : ["A0", "A1" , "A2" , "A3"],
                   "B" : ["B0", "B1", "B2", "B3"],
                   "C" : ["C0", "C1","C2" ,"C3"],
                   "D" : ["D0", "D1","D2" ,"D3"]
                    } , index = [0,1,2,3])

df5 = pd.DataFrame({"E" : ["E4", "E5" , "E6" , "E7"],
                   "F" : ["F4", "F5", "F6", "F7"],
                   "G" : ["G4", "G5","G6" ,"G7"],
                   "H" : ["D4", "D5","D6" ,"D7"]}
                    ,index = [0,1,2,3])

df6 = pd.DataFrame({"I" : ["I8", "I9" , "I10" , "I11"],
                   "J" : ["J8", "J9", "J10", "J11"],
                   "K" : ["K8", "K9","K10" ,"K11"],
                   "L" : ["L8", "L9","L10" ,"L11"]
                   }, index = [0,1,2,3])

```

```python
# STEP2 : concat을 이용한 데이터 결합
# STEP3 : parameter 변경

df = pd.concat([df4,df5,df6], axis = 1)
df
```
